```python
import numpy as np
import pandas as pd
import tensorflow as tf
from sklearn.model_selection import train_test_split

# Select input columns
input_cols = df.columns[:9]  # Exclude 'Fasilitas_ID' from input

# Map postal codes to unique indices
postal_codes = df['postal_code'].unique()
postal_code_to_index = {code: i for i, code in enumerate(postal_codes)}
df['postal_code_index'] = df['postal_code'].map(postal_code_to_index)

# Split the data into training and validation sets
train_data, val_data = train_test_split(df, test_size=0.2, random_state=42)

# Number of users and items
num_users = len(postal_codes)
num_items = len(df['Fasilitas_ID'].unique())

# Build a collaborative filtering model using matrix factorization with TensorFlow
embedding_dim = 10

# Input features
user_input = tf.keras.layers.Input(shape=(1,), name='postal_code')
other_inputs = [tf.keras.layers.Input(shape=(1,), name=col) for col in input_cols]

# Embedding layers for user and item
user_embedding = tf.keras.layers.Embedding(input_dim=num_users, output_dim=embedding_dim)(user_input)
other_embedding = [tf.keras.layers.Embedding(input_dim=unique_count, output_dim=embedding_dim)(input_) for unique_count, input_ in zip(df[input_cols].nunique(), other_inputs)]

# Flatten the embeddings
user_embedding_flat = tf.keras.layers.Flatten()(user_embedding)
other_embedding_flat = [tf.keras.layers.Flatten()(embedding) for embedding in other_embedding]

# Concatenate the flattened embeddings with other input features
concatenated = tf.keras.layers.Concatenate()([user_embedding_flat] + other_embedding_flat)

# Dense layers for prediction
dense1 = tf.keras.layers.Dense(50, activation='relu')(concatenated)
output = tf.keras.layers.Dense(num_items, activation='softmax')(dense1)  # Softmax for recommendation

model = tf.keras.Model(inputs=[user_input] + other_inputs, outputs=output)
model.compile(optimizer='adam', loss='categorical_crossentropy')  # Use categorical crossentropy for recommendation

# Prepare one-hot encoded labels for training
train_labels = tf.keras.utils.to_categorical(train_data['Fasilitas_ID'] - df['Fasilitas_ID'].min(), num_classes=num_items)
val_labels = tf.keras.utils.to_categorical(val_data['Fasilitas_ID'] - df['Fasilitas_ID'].min(), num_classes=num_items)

# Train the model with a validation split
model.fit([train_data['postal_code_index']] + [train_data[col] for col in input_cols], train_labels,
          epochs=10, batch_size=1, verbose=1,
          validation_data=([val_data['postal_code_index']] + [val_data[col] for col in input_cols], val_labels))

# Make predictions for the entire dataset
predictions = model.predict([df['postal_code_index']] + [df[col] for col in input_cols])

# Select the top-k recommended items for each user
top_k = 5  # Choose the number of top recommendations
recommended_ids = np.argsort(predictions, axis=1)[:, -top_k:]
```
